```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
tableau11 = [(31, 119, 180), (255, 127, 14),    
             (44, 160, 44), (214, 39, 40),   
             (148, 103, 189), (140, 86, 75),  
             (227, 119, 194), (127, 127, 127),   
             (188, 189, 34), (23, 190, 207), (70, 70, 70), (0, 255, 0), (0, 0, 255)]  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau11)):    
    r, g, b = tableau11[i]    
    tableau11[i] = (r / 255., g / 255., b / 255.)  
```

```python
df = pd.read_csv("dummy_mexp_150-2.csv", sep="\t")
```

```python
df = df.set_index('param')
df = df[df.columns[:-1:2]]
```

```python
plt.style.use(plt.style.available[1])

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
df.plot(color=tableau11, ax=ax)
ax.set_ylim(0.15, 1)

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('7cluster.png', bbox_inches='tight', dpi=400)
```
